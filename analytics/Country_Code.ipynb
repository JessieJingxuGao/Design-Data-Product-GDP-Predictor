```python
import pycountry

iso2=[]
iso3=[]
name=[]
official_name=[]

for item in list(pycountry.countries):
    try:
        iso2.append(item.alpha_2)
    except AttributeError:
        official_name.append(None)
    try:
        iso3.append(item.alpha_3)
    except AttributeError:
        official_name.append(None)
    try:
        official_name.append(item.official_name)
    except AttributeError:
        official_name.append(None)
    try:
        name.append(item.name)
    except AttributeError:
        name.append(None)
        
dict={"iso3": iso3,"iso2":iso2,"name":name,"official_name":official_name}

df=pd.DataFrame(dict)

df
```

```python
df.describe()
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('country_code.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
